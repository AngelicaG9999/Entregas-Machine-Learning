### **Ejercicio 2 (C = 1.0)**

Genere la siguiente imagen [resultado](https://drive.google.com/file/d/1FjNEt4uG1svgiDlMsmv3-CVR1qcciduh/view?usp=share_link).

**NOTA:** En la salida de la celda puede encontrar el tama√±o de la imagen final.

```
# Lea la imagen
# image = cv2.imread('Ex2.png')

# Realice aqu√≠ todas las operaciones que necesite. Puede a√±adir cuantas celdas de c√≥digo requiera
fc = 8 # n√∫mero de flas y columnas
tc = 50 # tama√±o de cada cuadro en pixeles
d = fc * tc
c = 1.0

img  = np.zeros((d, d), dtype=np.uint8)

for i in range(fc):
  for j in range(fc):

    if (i+j) % 2 == 0:
      color = 255
    else:
      color = 0

    fi = i * tc       #fila inicial
    ff = (i + 1) * tc #fila final
    ci = j * tc       #columna inicial
    cf = (j + 1) * tc #columna final

    img[fi:ff, ci:cf] = color * c

plt.imshow(img, cmap='gray')
plt.show()

# Muestre la imagen y guardela en drive usando opencv (consultar)
cv2.imwrite('/content/drive/MyDrive/Ex2.png', img)
```

### **Ejercicio 3 (C = 1.3)**

Genere la siguiente imagen [resultado](https://drive.google.com/file/d/1-9jWSwRb1ld_xN6iHTHVKT92bGat0I4s/view?usp=sharing).

**NOTA:** En la salida de la celda puede encontrar el tama√±o de la imagen final.

```
# Lea la imagen
img = cv2.imread('/content/drive/MyDrive/lena.jpg')

# Realice aqu√≠ todas las operaciones que necesite. Puede a√±adir cuantas celdas de c√≥digo requiera
roi = img[100:200,180:280]


f = 3 # filas
col = 4 # columnas
alto_roi, ancho_roi, canales = roi.shape
c = 1.3

img = np.zeros((alto_roi * f, ancho_roi * col, canales), dtype=np.uint8)

for i in range(f):
  for j in range(col):

    fi = i * alto_roi        #fila inicial
    ff = (i + 1) * alto_roi  #fila final
    ci = j * ancho_roi       #columna inicial
    cf = (j + 1) * ancho_roi #columna final

    img[fi:ff, ci:cf] = np.clip(roi * c, 0, 255).astype(np.uint8)

# Convert√≠ imagen de BGR a RGB, porque al subirla con cv2 se quedaba en BGR
img_rgb = img[..., ::-1]

# Muestre la imagen y guardela en drive usando opencv (consultar)
plt.imshow(img_rgb)
plt.show()
cv2.imwrite('/content/drive/MyDrive/lena-cuadrito.png', img)
```

### **Ejercicio 5 (C = 0.7)**

Genere la siguiente imagen [resultado](https://drive.google.com/file/d/10jxoa_muSn6bxsDBy7Vw88wz1oirMiRH/view?usp=drive_link)

```
# Lea la imagen
img = cv2.imread('/content/drive/MyDrive/lena.jpg')

# Realice aqu√≠ todas las operaciones que necesite. Puede a√±adir cuantas celdas de c√≥digo requiera
h, w, ch = img.shape # Alto, ancho y canal de la imagen
C = 0.7

# Crear una matriz en ceros de igual tama√±o a la imagen original como lienzo para poner la imagen espejo
lienzo = np.zeros((h, w, ch), dtype=np.uint8)

for y in range(h):          # Recorre filas es decir el alto de la imagen
    for x in range(w):      # Recorre columnas es decir ancho de la imagen
        for c in range(ch): # Recorre los canales de color (R, G, B)
            if y < h / 2:
                # Mitad superior de la imagen original
                img_espejo = int(img[y, x, c] * C)
            else:
                # Mitad inferior de la imagen original
                img_espejo = int(img[h - 1 - y, x, c] * C)

            lienzo[y, x, c] = img_espejo

# Convertir imagen de BGR a RGB, porque al subirla con cv2 se quedaba en BGR
img_rgb = lienzo[..., ::-1]

# Muestre la imagen y guardela en drive usando opencv (consultar)
plt.imshow(img_rgb)
plt.show()
cv2.imwrite('/content/drive/MyDrive/lena-espejo.png', lienzo)
```
